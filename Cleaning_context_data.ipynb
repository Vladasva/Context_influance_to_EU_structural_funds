```python
def clean_world_bank_data(df, name):
    df_1 = df[['REF_AREA_LABEL', 'UNIT_MEASURE', 'TIME_PERIOD', 'OBS_VALUE']]
    df_2 = df_1.loc[df_1['REF_AREA_LABEL'].isin(["Lithuania", "Latvia", "Estonia", "Denmark", "Sweden", "Finland", "Germany", "Poland"])]
    df_3 = df_2.loc[df_2['UNIT_MEASURE'].isin(["1_TO_7", "0_TO_10"])]
    df_4 = df_3.drop(["UNIT_MEASURE"], axis=1)
    df_5 = df_4.rename(columns = {'OBS_VALUE': name}) 
    return df_5
```

```python
def return_spefifc_country(df, name):
    df_1 = df.loc[df['REF_AREA_LABEL'].isin([name])]
    return df_1
```
