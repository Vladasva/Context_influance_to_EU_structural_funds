```python
import pandas as pd
pd.options.mode.copy_on_write = True
```

```python
def clean_world_bank_data(df):
    df_1 = df.loc[df['Country Name'].isin(["Denmark", "Germany", "Poland", "Lithuania", "Latvia", "Estonia", "Finland", "Sweden"])]
    df_2 = df_1.drop(["Country Code", "Indicator Code", "Indicator Name"], axis=1)
    df_3 = df_2.transpose()
    df_4 = df_3.rename({55:"Germany", 58:"Denmark", 71:"Estonia",
                                              75:"Finland", 143:"Lithuania", 145:"Latvia", 190:"Poland",
                                              223:"Sweden"}, axis=1)
    df_5 = df_4.tail(df_4.shape[0] - 1)
    df_6 = df_5.reset_index()
    df_7 = df_6.rename({"index":"TIME_PERIOD"}, axis=1)
    df_7 = df_7[df_7['TIME_PERIOD'].astype(str).str.isnumeric()]
    df_7['TIME_PERIOD'] = pd.to_numeric(df_7['TIME_PERIOD'])
    df_8 = df_7[df_7['TIME_PERIOD'].astype(str).str.isnumeric()]
    df_9 = df_8.set_index("TIME_PERIOD")
    return df_9
```

```python
def return_spefifc_country(df, name):
    df_1 = df.loc[df['REF_AREA_LABEL'].isin([name])]
    return df_1
```

```python
def clean_eurostat_data(df):
    df_1 = df.drop(["Unnamed: 2", "Unnamed: 4", "Unnamed: 6", "Unnamed: 8", "Unnamed: 10", "Unnamed: 12",
                              "Unnamed: 14", "Unnamed: 16", "Unnamed: 18", "Unnamed: 20", "Unnamed: 22", "Unnamed: 24", "Unnamed: 26", 
                              "Unnamed: 28", "Unnamed: 30", "Unnamed: 32", "Unnamed: 34", "Unnamed: 36", "Unnamed: 38"], axis=1)
    df_2 = df_1.transpose()
    df_3 = df_2.drop([0, 9, 10, 11, 12, 13], axis=1)
    df_4 = df_3.rename({2:"Germany", 1:"Denmark", 3:"Estonia", 7:"Finland", 5:"Lithuania", 4:"Latvia", 6:"Poland", 8:"Sweden"}, axis=1)
    df_5 = df_4.tail(df_4.shape[0] - 1)
    df_6 = df_5.reset_index()
    df_7 = df_6.rename({'index':'TIME_PERIOD'}, axis=1)
    df_7 = df_7[df_7['TIME_PERIOD'].astype(str).str.isnumeric()]
    df_7['TIME_PERIOD'] = pd.to_numeric(df_7['TIME_PERIOD'])
    df_8 = df_7[df_7['TIME_PERIOD'].astype(str).str.isnumeric()]
    df_9 = df_8.set_index("TIME_PERIOD")
    return df_9

    
```

```python
def clean_eurostat_data_1(df):
    df_1 = df.drop(["Unnamed: 2", "Unnamed: 4", "Unnamed: 6", "Unnamed: 8", "Unnamed: 10", "Unnamed: 12",
                              "Unnamed: 14", "Unnamed: 16", "Unnamed: 18", "Unnamed: 20", "Unnamed: 22", "Unnamed: 24", "Unnamed: 26", 
                              "Unnamed: 28", "Unnamed: 30", "Unnamed: 32", "Unnamed: 34"], axis=1)
    df_2 = df_1.transpose()
    df_3 = df_2.drop([0, 9, 10, 11, 12, 13], axis=1)
    df_4 = df_3.rename({2:"Germany", 1:"Denmark", 3:"Estonia", 7:"Finland", 5:"Lithuania", 4:"Latvia", 6:"Poland", 8:"Sweden"}, axis=1)
    df_5 = df_4.tail(df_4.shape[0] - 1)
    df_6 = df_5.reset_index()
    df_7 = df_6.rename({'index':'TIME_PERIOD'}, axis=1)
    df_7 = df_7[df_7['TIME_PERIOD'].astype(str).str.isnumeric()]
    df_7['TIME_PERIOD'] = pd.to_numeric(df_7['TIME_PERIOD'])
    df_8 = df_7[df_7['TIME_PERIOD'].astype(str).str.isnumeric()]
    df_9 = df_8.set_index("TIME_PERIOD")
    return df_9
```
